
Model

$$ \hat{\mathbf{Y}} = \phi(\mathbf{X}; \boldsymbol{\Theta}) = \boldsymbol{\Theta}^T \mathbf{X} + \mathbf{\Epsilon}
 $$


Loss function
$$ Q(\boldsymbol{\Theta}) = \frac{1}{N} \sum_{i=1}^N e_i^2 $$

<br>

$$ e_i = y_i - \hat{y}_i  \ \ \ \ \text{for} \ \ i = 1, \dots, N $$



$$ \mathbf{\Epsilon} =
\begin{bmatrix}
e_1 & \dots & e_N
\end{bmatrix} =
\begin{bmatrix}
y_1 - \hat{y}_1 & \dots & y_N - \hat{y}_N   
\end{bmatrix}
$$

<br>

$$ (\text{WLOG}) \ \ \frac{1}{N} \sum_{i=1}^N e_i^2 = \sum_{i=1}^N e_i^2 $$

$$
\sum_{i=1}^N e_i^2 
= \mathbf{E}\mathbf{E}^T
= (\mathbf{Y} - \hat{\mathbf{Y}})(\mathbf{Y} - \hat{\mathbf{Y}})^T
= (\mathbf{Y} - \boldsymbol{\Theta}^T \mathbf{X})(\mathbf{Y} - \boldsymbol{\Theta}^T \mathbf{X})^T
$$

$$
= (\mathbf{Y} - \boldsymbol{\Theta}^T \mathbf{X})(\mathbf{Y}^T - \mathbf{X}^T \boldsymbol{\Theta})
= \mathbf{Y}\mathbf{Y}^T - \mathbf{Y}\mathbf{X}^T\boldsymbol{\Theta}
- \boldsymbol{\Theta}^T \mathbf{X}\mathbf{Y}^T
+ \boldsymbol{\Theta}^T \mathbf{X}\mathbf{X}^T \boldsymbol{\Theta}
$$

<br>

$$ 
\frac{\partial Q}{\partial \boldsymbol{\Theta}} = 0
$$

$$ 
\frac{\partial}{\partial \boldsymbol{\Theta}}
\left(
\mathbf{Y}\mathbf{Y}^T
- \mathbf{Y}\mathbf{X}^T\boldsymbol{\Theta}
- \boldsymbol{\Theta}^T \mathbf{X}\mathbf{Y}^T
+ \boldsymbol{\Theta}^T \mathbf{X}\mathbf{X}^T \boldsymbol{\Theta}
\right) = 0
$$

$$
- \mathbf{X}\mathbf{Y}^T - \mathbf{X}\mathbf{Y}^T
+ 2\mathbf{X}\mathbf{X}^T \boldsymbol{\Theta} = 0
$$


$$
\mathbf{X}\mathbf{X}^T \boldsymbol{\Theta} = \mathbf{X}\mathbf{Y}^T
$$

$$
\boldsymbol{\Theta} = (\mathbf{X}\mathbf{X}^T)^{-1}\mathbf{X}\mathbf{Y}^T
$$

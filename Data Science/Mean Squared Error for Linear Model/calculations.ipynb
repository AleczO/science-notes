
Model

$$ \hat{Y} = \phi(X; \Theta) = \Theta^T X + E $$


Loss function
$$ Q(\Theta) = \frac{1}{N} \sum_{i=1}^N e_i^2 $$

<br>

$$ e_i = y_i - \hat{y}_i  \ \ \ \ \text{for} \ \ i = 1, \dots, N $$



$$ E = 
\begin{bmatrix}
e_1 & \dots & e_N
\end{bmatrix} =
\begin{bmatrix}
y_1 - \hat{y}_1 & \dots & y_N - \hat{y}_N   
\end{bmatrix}
$$

<br>

$$ (\text{WLOG}) \ \ \frac{1}{N} \sum_{i=1}^N e_i^2 = \sum_{i=1}^N e_i^2 $$

$$
\sum_{i=1}^N e_i^2 = EE^T = (Y - \hat{Y})(Y - \hat{Y})^T = (Y - \Theta^T X)(Y - \Theta^T X)^T = 
$$

$$
= (Y - \Theta^T X)(Y^T - X^T\Theta) = YY^T - YX^T\Theta - \Theta^T XY^T + \Theta X X^T \Theta
$$

<br>

$$ 
\frac{\partial Q}{\partial \Theta} = 0
$$

$$ 
\frac{\partial}{\partial \Theta}(YY^T - YX^T\Theta - \Theta^T XY^T + \Theta X X^T \Theta) = 0
$$

$$
-XY^T - XY^T + 2XX^T \Theta = 0
$$


$$
XX^T \Theta = XY^T
$$

$$
\Theta = (XX^T)^{-1}XY^T
$$

$$ \mathbf{A} \mathbf{v} = \lambda \mathbf{v} $$

for corresponding pair $(\text{eigenvector} \ \ \mathbf{v}, \ \text{eigenvalue} \ \ \lambda)$

<br>

$$ \mathbf{V} = 
\begin{bmatrix}
\mathbf{v}_1 & \dots & \mathbf{v}_n
\end{bmatrix}
$$


$$ 
\mathbf{D} = \text{diag} (\lambda_1, \dots, \lambda_n)
$$

<br>

$$
\mathbf{A} \mathbf{V} = \mathbf{V} \mathbf{D}
$$

$$
\mathbf{A} = \mathbf{V} \mathbf{D} \mathbf{V}^{-1}
$$


It can be easily verified

$$

\mathbf{A}\mathbf{V} =

\mathbf{A} 
\begin{bmatrix}
\mathbf{v}_{1} & \dots & \mathbf{v}_{n} \\
\end{bmatrix} =

\begin{bmatrix}
\mathbf{A} \mathbf{v}_{1} & \dots & \mathbf{A} \mathbf{v}_{n} \\
\end{bmatrix}

$$

$$
\mathbf{V} \mathbf{D} = \begin{bmatrix}
\mathbf{v}_{1} & \dots & \mathbf{v}_{n} \\
\end{bmatrix}

\begin{bmatrix}
\lambda_1 & & 0 \\
& \ddots & \\
0 & & \lambda_n 
\end{bmatrix} =

\mathbf{V} \mathbf{D} = \begin{bmatrix}
\lambda_1\mathbf{v}_{1} & \dots & \lambda_n \mathbf{v}_{n} \\
\end{bmatrix}
$$

<br>

$$ \mathbf{A} \mathbf{v}_i = \lambda_i \mathbf{v}_i \ \ \ \forall i \in \{1, \ \dots, n\} $$

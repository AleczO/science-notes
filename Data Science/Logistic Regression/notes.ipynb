
$$ \mathbf{x}^{(i)} = \{ {x}^{(i)}_1, {x}^{(i)}_2, \dots, {x}^{(i)}_M \} \in \mathbb{R}^M  \ \ \ , i \in \{1, \dots, n \} $$

$$ y^{(i)} \in \{0, 1\} $$

<br>

$$ 
\begin{align*}
P(y^{(i)} = 1 | \mathbf{x}^{(i)}, \mathbf{w}) &= p^{(i)}  = \sigma(\mathbf{\mathbf{w}}^T \mathbf{x}^{(i)}) = \frac{1}{1 + \exp{(-\mathbf{w}^T \mathbf{x}^{(i)})}} \\ \\
P(y^{(i)} = 0 | \mathbf{x}^{(i)}, \mathbf{w}) &= 1 - p^{(i)} 
\end{align*} 
$$

<br>

$$ P(y^{(i)} | \mathbf{x}^{(i)}, \mathbf{w}) = (p^{(i)})^{y^{(i)}} (1 - p^{(i)})^{1 - y^{(i)}} $$

$$ 
\mathcal{L} = \prod_{i = 1}^n P(y^{(i)} | \mathbf{x}^{(i)}, \mathbf{w}) =  
\prod_{i = 1}^n \bigg[ \frac{1}{1 + \exp{(-\mathbf{w}^T \mathbf{x}^{(i)})}} \bigg]^{y^{(i)}}
                \bigg[1 - \frac{1}{1 + \exp{(-\mathbf{w}^T \mathbf{x}^{(i)})}} \bigg]^{1 - y^{(i)}}
$$

<br>





# MAP Estimator


$$ \text{Bayes Rule} $$

$$ p(\mathbf{w} | \mathbf{X}, \mathbf{y}) = \frac{p(\mathbf{y} | \mathbf{X, \mathbf{w}}) p(\mathbf{w})}{p(\mathbf{\mathbf{y}} | \mathbf{X})} \propto p(\mathbf{y} | \mathbf{X, \mathbf{w}}) p(\mathbf{w}) $$


$$
p(\mathbf{y} | \mathbf{X, \mathbf{w}}) p(\mathbf{w})  = \prod_{i = 1}^n \bigg[ \frac{1}{1 + \exp{(-\mathbf{w}^T \mathbf{x}^{(i)})}} \bigg]^{y^{(i)}}
                \bigg[1 - \frac{1}{1 + \exp{(-\mathbf{w}^T \mathbf{x}^{(i)})}} \bigg]^{1 - y^{(i)}} \exp{\bigg(-\frac{1}{2}\mathbf{w}^T \mathbf{w} \bigg)}
$$

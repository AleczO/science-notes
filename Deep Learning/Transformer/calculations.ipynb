# Scaling

---
$$ 
\begin{align*}
\text{cov}(\mathbf{a}^T \mathbf{b}) = \mathbb{E}[(\mathbf{a}^T\mathbf{b})^2] &= \mathbb{E}\bigg[ \bigg( \sum_{d=1}^D a_d b_d \bigg)^2 \bigg] \\

&= \mathbb{E}\bigg[  \sum_{d=1}^D ( a_d b_d )^2 + \sum_{d=1}^D \sum_{\substack{e = 1 \\ e \neq d}}^{D} a_d b_d a_e b_e \bigg] \\
\\

& (*) \ \ \ \ \mathbb{E}[\alpha\mathbf{x} + \beta \mathbf{y}] = \alpha \mathbb{E} [\mathbf{x}] + \beta \mathbb{E}[\mathbf{y}]  \ \ \ \ \\
\\

&=\sum_{d=1}^D \mathbb{E}\big[( a_d b_d )^2\big] + \sum_{d=1}^D \sum_{\substack{e = 1 \\ e \neq d}}^{D} \mathbb{E}[a_d b_d a_e b_e] \\

&=\sum_{d=1}^D \bigg(  \mathbb{E}\big[( a_d b_d )^2\big] +  \sum_{\substack{e = 1 \\ e \neq d}}^{D} \mathbb{E}[a_d b_d a_e b_e]  \bigg) \\

\\

& (*) \ \ \ \ \mathbf{a} \perp\!\!\!\perp \mathbf{b} \ \ \ \land \ \ \ \mathbf{a}, \mathbf{b} \sim \mathcal{N}(\mathbf{0}, \mathbf{I}) \\

\\

&=\sum_{d=1}^D \bigg(  \mathbb{E}\big[ a_d^2\big]\mathbb{E}\big[ b_d^2\big]  +  \sum_{\substack{e = 1 \\ e \neq d}}^{D} \mathbb{E}[a_d] \mathbb{E}[b_d] \mathbb{E}[a_e] \mathbb{E}[b_e]  \bigg) \\

&=\sum_{d=1}^D \bigg(1\cdot 1 + \sum_{\substack{e = 1 \\ e \neq d}}^{D} 0 \cdot 0 \cdot 0 \cdot 0 \bigg) \\

&=D
\end{align*}
$$
